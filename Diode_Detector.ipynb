### Double Diode Model

### Diode Voltage
The diode voltage $V_D$ is reduced from the applied bias $V_{bias}$ due to series resistance $R_s$. We have $V_D = V_{bias} - J_D R_s$. 


### Dark Current Density


\begin{equation}
J_{dark} = \sum_{n} {J_{n}\left[\exp{\left(\frac{qV_D}{nkT}\right)} - 1\right]} + \frac{V_D}{R_{SH}}
\end{equation}

### Photo-current
\begin{gather}
J_{light} = q\eta Q_{det} \\
I_{light} = A_{opt} J_{light}
\end{gather}

### Total Diode Current
\begin{align}
J_D &= J_{dark} - J_{light} \\
&= \sum_{n} {J_{n}\left[\exp{\left(\frac{qV_D}{nkT}\right)} - 1\right]} + \frac{V_D}{R_{SH}} - q\eta Q_{det}
\end{align}

### Diode Resistance-Area Product
\begin{align}
\frac{1}{RA_{D}} &= \frac{dJ_{D}}{dV_{D}} \\
& = \sum_{n}{J_{n} \left[ \frac{q}{nkT} \exp{\left(\frac{qV_{D}}{nkT}\right)} \right]} + \frac{1}{R_{SH}}
\end{align}
